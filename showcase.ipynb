```python
import imageaipackage as iap
import numpy as np
import cv2
from matplotlib import pyplot as plt
import demo
import loader
```

```python
img1 = iap.image_path_to_np_array("unit_test_images/TestImg1.jpg")
img2 = iap.image_path_to_np_array("unit_test_images/TestImg2.jpg")
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.imshow(img1_rgb)
plt.subplot(1,2,2)
plt.imshow(img2_rgb)
```

```python
new_img = iap.crop(img1, 170)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
```

```python
new_img = iap.resize(img1, 170)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
```

```python
new_img = iap.region_grow(img2, 10)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
```

```python
new_img = iap.edge_detection(img1)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
```

```python
# Blur
new_img = iap.blur(img1, kernel_size=5)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Blurred Image")
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Bilateral Blur
new_img = iap.bilateral_blur(img1, diameter=9, sigma_color=75, sigma_space=75)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Bilateral Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Gaussian Blur
new_img = iap.gaussian_blur(img1, kernel_size=5, sigma=3)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Median Blur
new_img = iap.median_blur(img1, kernel_size=5)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Median Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Laplacian
new_img = iap.laplacian(img1, kernel_size=3)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Laplacian Image")
plt.axis('off')
plt.show()
```

```python
# Custom Kernel Blur
kernel = np.array([[1, 1, 1], [1, -7, 1], [1, 1, 1]])
new_img = iap.custom_kernel_blur(img1, kernel)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Custom Kernel Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Sharpen
new_img = iap.sharpen(img1, sharpness=9.0)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Sharpened Image")
plt.axis('off')
plt.show()
```

```python
# Adaptive Thresholding
new_img = iap.adaptive_thresholding(img1, block_size=11, const_c=2)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Adaptive Thresholded Image")
plt.axis('off')
plt.show()
```

```python
# Otsu's Binarization Thresholding
new_img = iap.otsus_binarization_thresholding(img1)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Otsu's Binarization Thresholded Image")
plt.axis('off')
plt.show()
```

```python
# Edge Detection (Canny)
new_img = iap.edge_detection(img1, min_val=100, max_val=200)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Edge Detection (Canny) Image")
plt.axis('off')
plt.show()
```

```python
# Watershed
new_img = iap.watershed(img1)
new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_rgb)
plt.title("Watershed Segmentation Image")
plt.axis('off')
plt.show()
```
